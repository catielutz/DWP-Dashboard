```python
import pandas as pd
df_2010 = pd.read_csv("../full csvs/Guttmacher_Clinic_Census_County_2010.csv",
                   dtype={"FIPS": str})

# add leading zero to county FIPS codes where needed
df_2010_FIPS = df_2010["FIPS"].astype("str").str.zfill(5)
compFIPS = pd.DataFrame(df_2010_FIPS)

# insert updated FIPS info back to orig df
df_2010["FIPS"] = compFIPS["FIPS"].values
df_2010
```

```python
# print column names to select what to pull
df_2010.columns
```

```python
# pull state, FIPS, county, and total public funded clinics columns only.
county2010 = df_2010[['State', 'FIPS', 'County', 'Total publicly funded clinics']]
county2010.head()
```

```python
# rename column to remove spaces
county2010 = county2010.rename(columns={'Total publicly funded clinics': 'Pub_Fund_Clinics'})
county2010
```

```python
import json

# convert df to json and export file
county2010.to_json(r'../2010_clinics.json', orient="table")
```
